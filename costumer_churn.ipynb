```python
ax = sns.countplot(x= "Churn", data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Churn")
plt.show()
```

```python
plt.figure(figsize=(5,5))

gb = df.groupby("Churn").agg({"Churn":"count"})
gb
plt.pie(gb['Churn'], labels= gb.index, autopct="%1.2f%%")
plt.title("Percent of Churned Distribution", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x= "gender", data=df, hue="Churn")
plt.legend(loc="upper right")
plt.title("Count of Gender")
plt.show()
```

```python
plt.figure(figsize=(5,5))
ax =  sns.countplot(x = "SeniorCitizen", data= df)
ax.bar_label(ax.containers[0])
plt.legend(loc="upper right")
plt.title("Count of Senior Citizen")
plt.show()
```

```python
senior_churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate percentages
senior_churn_percentages = senior_churn_counts.apply(lambda x: x / x.sum() * 100, axis=1)

# Create stacked bar chart
ax = senior_churn_percentages.plot(kind='bar', stacked=True, figsize=(6, 5))

# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.title("Churn Percentage by Senior Citizen Status")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.legend(title="Churn",bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
# comparative a greater percentage of people in senior citizen category have churned
```

```python
sns.histplot(x= "tenure", data=df, bins=50, hue="Churn")
plt.title("Count of Tenure")
plt.show()
# people who have used our services for long time have stayed and people who have used our services 1 or 2 month have churned.
```

```python
plt.figure(figsize=(5,5))
ax =  sns.countplot(x = "Contract", data= df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.legend(loc="upper right")

plt.title("Count of Customer by Contract ")
plt.show()
# people who have month to month contract are likely to churn then from those who have 1 or 2 years or contract
```

```python
cols_to_plot = ['PhoneService', 'MultipleLines', 'InternetService',
                'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport', 'StreamingTV', 'StreamingMovies']

num_cols = len(cols_to_plot)
num_rows = (num_cols + 2)
  # Calculate number of rows needed (3 columns per row)

fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows * 3))
axes = axes.flatten() # Flatten the 2D array of axes to easily iterate

for i, col in enumerate(cols_to_plot):
    sns.countplot(x=col, data=df, hue="Churn", ax=axes[i])
    axes[i].set_title(f'Count of {col}')
    axes[i].legend(loc="upper right")

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# The majority of customers who do not churm tend to have services like PhoneService, InternetService(Particularly DSL), and OnlineSecurity enable. For services like OnlineBackup
# TechSupport, and StreamimgTV, Churm rates are noticeably higher when these services are not used or are unavailable.
```
